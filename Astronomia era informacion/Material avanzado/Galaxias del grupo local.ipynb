```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Partiremos graficando las galaxias del grupo local en un grafico tipo Mollweide

```python
# EJECUTE ESTE CODIGO
# Definimos la funcion plot_mwd para hacer graficos en una proyecciÃ³n del tipo Mollweide
import collections

def flatten(x):
    if isinstance(x, collections.Iterable):
        return [a for i in x for a in flatten(i)]
    else:
        return [x]

def plot_mwd(l,b,org=0,title='Mollweide projection', projection='mollweide', label='', label_id=''):
    x = np.remainder(l+360-org,360) # shift RA values
    ind = x>180
    x[ind] -=360    # scale conversion to [-180, 180]
    x=np.radians(-x)    # reverse the scale: East to the left
    y=np.radians(b)
    tick_labels = np.array([150, 120, 90, 60, 30, 0, 330, 300, 270, 240, 210])
    tick_labels = np.remainder(tick_labels+360+org,360)

    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111, projection=projection, axisbg ='LightCyan')
    ax.scatter(x, y, marker='o', s=15, facecolor='red', edgecolor='', alpha=0.7)  # convert degrees to radians
    ax.set_xticklabels(tick_labels, alpha=0.4)     # we add the scale on the x axis
    ax.set_title(title)
    ax.title.set_fontsize(15)
    ax.set_xlabel("Longitud")
    ax.xaxis.label.set_fontsize(12)
    ax.set_ylabel("Latitud")
    ax.yaxis.label.set_fontsize(12)
    
    if type(label_id)==int:
        label_id=(label_id,)
    elif type(label_id)==list:
        label_id=flatten(label_id)

    try:
        for i in range(len(label_id)): 
            ax.annotate(unicode(label[label_id[i]], "utf-8"), xy=(x[label_id[i]], y[label_id[i]]), xytext=(x[label_id[i]]+2e-1, y[label_id[i]]+1e-1), fontsize=10, arrowprops=dict(facecolor='black', shrink=0.2, width=0.5, frac=0.2, headwidth=4))
    except:
        print ''
    ax.grid(True)
```

```python
# Cargamos catalogo de las galaxias del grupo local
# La distancia y el diametro esta en unidades de kpc

cat_url='../data/catalogo_grupo_local.csv' #'https://raw.githubusercontent.com/rpmunoz/PentaUC/master/Astronomia%20era%20informacion/data/catalogo_grupo_local.csv'
cat_data=pd.read_csv(cat_url,header=0,sep=',')
cat_data.head()
```

```python
# Ahora ordenaremos las galaxias usando el diametro de mayor a menor
cat_data.sort_values('diametro',ascending=False)
```

```python
# Anote el ID de la galaxia Andromeda
id_vialactea=0
cat_data.loc[id_vialactea]
```

```python
# Haga un grafico del tipo Mollweide de como se distribuyen las galaxias en el cielo. Use el id de la via lactea
plot_mwd(cat_data['l'], cat_data['b'], label=cat_data['nombre'], label_id=id_vialactea)
```

```python
# Anote el ID de la galaxia Andromeda
id_andromeda=36
cat_data.loc[id_andromeda]
```

```python
# Haga un grafico del tipo Mollweide de como se distribuyen las galaxias en el cielo. Use Complete label_id y label_text
plot_mwd(cat_data['l'], cat_data['b'], label=cat_data['nombre'], label_id=id_andromeda)
```

```python
# Anote el ID de la pequeÃ±a y gran nube de magallanes
id_magallanes=[4,5]
cat_data.loc[id_magallanes]
```

```python
# Grafique e indique la posicion de las nubes de Magallanes 
ax=plot_mwd(cat_data['l'],cat_data['b'], label=cat_data['nombre'], label_id=id_magallanes)
```

```python
# Grafique e indique la posicion de la Via Lactea, Nubes de Magallanes y Andromeda
ax=plot_mwd(cat_data['l'],cat_data['b'], label=cat_data['nombre'], label_id=[id_vialactea,id_magallanes,id_andromeda])
```

## Ahora graficaremos las galaxias del grupo local en un grafico 3D

```python
# EJECUTE EL SIGUIENTE CODIGO

import collections
from astropy.coordinates import SkyCoord, Distance
from astropy import units as u
import pylab
from mpl_toolkits.mplot3d import Axes3D, proj3d

def flatten(x):
    if isinstance(x, collections.Iterable):
        return [a for i in x for a in flatten(i)]
    else:
        return [x]

def plot_sphere_3d(x, y, z, r, label='', label_id='', plot_range=(-10,10), xrange=(-10,10),yrange=(-10,10),zrange=(-10,10)):

    xrange=yrange=zrange=plot_range
    
    fig=plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlim(xrange)
    ax.set_ylim(yrange)
    ax.set_zlim(zrange)
    ax.set_xlabel('X (kpc)')
    ax.set_ylabel('Y (kpc)')
    ax.set_zlabel('Z (kpc)')
    
    for (xi,yi,zi,ri) in zip(x,y,z,r):
        u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
        xs=np.cos(u)*np.sin(v)
        ys=np.sin(u)*np.sin(v)
        zs=np.cos(v)
        # shift and scale sphere
        xs = ri*xs + xi
        ys = ri*ys + yi
        zs = ri*zs + zi
        
        ax.plot_wireframe(xs, ys, zs, color="r")
    
    if type(label_id)==int:
        label_id=(label_id,)
    elif type(label_id)==list:
        label_id=flatten(label_id)
        
    try:
        for i in range(len(label_id)):
#            print x[label_id[i]].value,y[label_id[i]].value,z[label_id[i]].value
            x2, y2, _ = proj3d.proj_transform(x[label_id[i]].value,y[label_id[i]].value,z[label_id[i]].value, ax.get_proj())
#            print x2, y2, label[label_id[i]]
            pylab.annotate(label[label_id[i]], xy = (x2, y2), xytext = (10, 10),
            textcoords = 'offset points', ha = 'left', va = 'bottom',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
    except:
        print ''
```

```python
# Convertimos las coordenadas galacticas en coordenadas cartesianas para desplegarlas en un grafico 3D
coo = SkyCoord(l=np.array(cat_data['l'])*u.degree, b=np.array(cat_data['b'])*u.degree, distance=np.array(cat_data['distancia'])*u.kpc, frame='galactic')
x=coo.cartesian.x
y=coo.cartesian.y
z=coo.cartesian.z
r=np.array(cat_data['diametro'])/2*u.kpc
```

```python
plot_sphere_3d(x,y,z,r, label=cat_data['nombre'], label_id=id_vialactea)
```

```python
plot_sphere_3d(x,y,z,r, label=cat_data['nombre'], label_id=[id_vialactea,id_magallanes], plot_range=(-40,40))
```

```python
plot_sphere_3d(x,y,z,r, label=cat_data['nombre'], label_id=[id_vialactea,id_magallanes,id_andromeda], plot_range=(-40,40))
```

```python
plot_sphere_3d(x,y,z,r, label=cat_data['nombre'], label_id=[id_vialactea,id_magallanes,id_andromeda], plot_range=(-80,80))
```
